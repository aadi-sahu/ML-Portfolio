```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
import joblib
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("gym_churn.csv")

X = df.drop("Churn", axis=1)
Y = df["Churn"]

cat_features = ['gender', 'Near_Location', 'Partner', 'Promo_friends', 'Phone', 'Contract_period', 'Group_visits']
num_features = ['Age', 'Avg_additional_charges_total', 'Month_to_end_contract', 'Lifetime', 'Avg_class_frequency_total', 'Avg_class_frequency_current_month']

preprocess = ColumnTransformer([
  ('num', StandardScaler(), num_features),
  ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features)
])

model = Pipeline([
  ('prep', preprocess), 
  ('clf', RandomForestClassifier(n_estimators=100, random_state=42))
])

model.fit(X, Y)

test_members = pd.DataFrame([
    [0, 1, 0, 0, 1, 1, 0, 22, 50.5, 1, 1, 0.5, 0.2],
    [1, 1, 1, 1, 1, 12, 1, 37, 200.2, 10, 15, 3.0, 2.9] 
], columns=X.columns)

predictions = model.predict(test_members)
probabilities = model.predict_proba(test_members)

for i, (pred, prob) in enumerate(zip(predictions, probabilities)):
    result = "QUIT" if pred == 1 else "STAY"
    chance = prob[1]
    print(f"Member {i+1} Analysis:")
    print(f" - Prediction: Customer will {result}")
    print(f" - Probability of Leaving: {chance:.2%}")
    print("-" * 30)
```
