```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Customer_Churn_Modelling.csv')

X = df[['Age']].values  
y = df['Balance'].values

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='skyblue', alpha=0.3, label='Data Points', s=10) 
plt.plot(X, y_pred, color='red', linewidth=3, label='Best Fit Line')     


plt.title('Linear Regression: Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.savefig('best_fit_line.png')
plt.show()

print(f"Intercept (b): {model.intercept_}")
print(f"Slope (m): {model.coef_[0]}")

from sklearn.metrics import r2_score
r2 = r2_score(y, y_pred)
print(f"R-squared Score: {r2}")



features = ['Age', 'CreditScore', 'Tenure', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']
X_multi = df[features]
y_balance = df['Balance']


multi_model = LinearRegression()
multi_model.fit(X_multi, y_balance)

print(f"New R-squared Score: {multi_model.score(X_multi, y_balance)}")

```
